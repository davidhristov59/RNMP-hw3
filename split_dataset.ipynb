```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./data/diabetes_012_health_indicators_BRFSS2015.csv')
df.head()
```


Податочното множество (од датотеката diabetes _ binary _ health _ indicators _ BRFSS2015.csv) потребно е да го поделите случајно во 2 датотеки:

offline.csv (80% од големината на оригиналното множество) \
online.csv (20% од големината на оригиналното множество)

```python
offline_df, online_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
offline_df.to_csv('./data/offline.csv', index=False)
online_df.to_csv('./data/online.csv', index=False)
```
